```
import healpy as hp
from ccgpack import sky2patch
import numpy as np
import pylab as plt
```

```
#dv = hp.read_map('./pla.esac.esa.int/pla-sl/data-action?SIMULATED_MAP.SIMULATED_MAP_OID=44065' , field=0 ,nest=1 , hdu=1 , verbose=1 )
dv = hp.read_map('./data/Smica-separated_cmb_0000.fits' , field=0 ,nest=1 , hdu=1 , verbose=1 )
```

```
hp.mollview(dv*1e6 , title='comp-sep' , nest= 1 , min=-300, max=300 , cmap='jet' )
```

```
dv8 = sky2patch(dv ,npatch=8)
```

```
fig, ax = plt.subplots(1, 1 , figsize=(5,5))
ax.imshow(dv8[0], cmap='jet')
ax.set_title('comp-sep' , size=16)
```
